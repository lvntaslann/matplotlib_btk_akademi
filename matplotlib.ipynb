```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# yazdığımız grafikleri kolayca gösterebilmek için
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
# "g" ifadesi grafiğin rengi için 
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun yaşa göre değişim grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
# "g--" , "g*-" kullanımları da yapılabilir"
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.xlabel("numpyDizisi1")
plt.ylabel("numpyDizisi2")
plt.title("numpyDizisi1 ve numpyDizisi2 karşılaştırması")
```

```python
# tek sıra 2 grafik ve 1.yi gösteriyorum anlamına geliyor
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra 2 kolon olacak ve 2.grafik
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
```

```python
# [sol kenar,alt kenar,genişlik,yükseklik]
benimFigure = plt.figure() #boş bir axis
figureAxes = benimFigure.add_axes([0.2, 0.2, 0.6, 0.6])
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes.set_xlabel("Xekseni")
figureAxes.set_ylabel("Yekseni")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.8,0.8])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"r--")
eksen1.set_xlabel("Xekseni")
eksen1.set_ylabel("Yekseni")
eksen1.set_title("Ana grafik başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b*-")
eksen2.set_xlabel("Xekseni küçük")
eksen2.set_ylabel("Yekseni küçük")
eksen2.set_title("Küçük grafik başlığı")
```

```python
#boş çalışınca eksen oluşturur yani bir tuple döndürür
(benimFigure,benimEksen) = plt.subplots()
```

```python
#boş çalışınca eksen oluşturur yani bir tuple döndürür
(benimFigure,benimEksenler) = plt.subplots(nrows = 1, ncols = 2)
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")

#araya mesafe koymak için
plt.tight_layout()
```

```python
# dpi seçeneği ile grafiğin kalitesi ayarlanabilir
# yeniFigure.savefig("benimfigur.png", dpi=200) ile görüntü kalitesi sonradan değiştirilip kaydedilebilir

yeniFigure = plt.figure()

# figure olustururken figsize=(3,3) gibi komutla da figure boyutları belirlenebilir
yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi **2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi **3")
yeniEksen.legend() # loc=1 gibi belirtilerek nerede bulunacağı belirtilebilir
```

## matplotlib görsellik 

```python
(benimFigur , benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color = "#3A95A8") #alpha komudu ile saydamlık verilebilir
benimEksen.plot(numpyDizisi2,numpyDizisi1,color = "#C96F23")
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1+2, color = "yellow", alpha = 0.7, linewidth = 5.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1+3, color = "red",alpha = 0.7, linewidth = 6.0)

yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 4 ,color = "green", linestyle = "--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 5 ,color = "purple", linestyle = ":")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 6 ,color = "blue", linestyle = "-.")

#marker , markersize , markercolor ile verilerin gösterimini detaylandırabiliriz
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 7 ,color = "#000000", linestyle = "-.",marker = "o",markersize = 4 ,markerfacecolor = "red")
```

## Pie chart

```python
meyveler = ["elma" , "portakal", "muz", "çilek"]
size = [25,35,20,20]
colors = ["red","orange","yellow","pink"]

# startangle grafiğin açısı için
plt.pie(size,labels=meyveler,colors = colors, startangle = 140 )
plt.title("Meyve tüketimi")
```

## Bar chart

```python
kategori = ["A","B","C","D"]
degerler = [3,7,5,4]

plt.bar(kategori,degerler,color = "blue")
plt.title("Kategorik veri dağılımı")
plt.xlabel("kategoriler")
plt.ylabel("Değerler")
```

## contour plots

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X,Y,Z)
plt.title("kontur grafiği")
```

## heatmaps

```python
data = np.random.rand(10,10)
# cmap her bir değer bir renge dönüştürmek için kullanılır
# interpolation ile pikseller arasındaki değerlerin nasıl hesaplanacağını belirler
# nearest diyerek bir önceki komşusuna göre belirleme işlemi yapılır
plt.imshow(data,cmap = "hot", interpolation = "nearest")
plt.colorbar() # renk skalası
plt.title("Isı haritası")
```
